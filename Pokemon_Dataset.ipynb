```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.txt', delimiter= "\t")

#The line below is an example to add new columns
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] +df['Sp. Def'] + df['Speed']
#to drop a column
#df = df.drop(columns= ['Total'])

#shorter way in making the total column
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

#rearranging columns
cols = list(df.columns.values)
df= df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(5)
```

# Aggregate Statistics (Group By)

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv')
#use groupby to group certain data types
df.groupby(['Type 1']).mean()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv')

df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
#can use .sum(), .mean(),.count()

df['count']= 1

df.groupby(['Type 1']).count()['count']
```

```python
#can add more parameters of that data 
df['count']= 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```
